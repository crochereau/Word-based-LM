```python
import numpy as np
import string
import pandas as pd
```

```python
# load doc into memory
def load_doc(filename):
    # open the file as read only
    file = open(filename, 'r')
    # read all text
    text = file.read()
    # close the file
    file.close()
    return text
```

```python
# load document
file = 'sentences.txt'
doc = load_doc(file)
print(doc[:100])
```

```python
def prepare_sentences(doc):
    lines = doc.split(" ")
    #print(lines[0])
    #print(len(lines))
    #print(lines[:5])
    
    #table = str.maketrans('','', string.punctuation)
    #tokens = [word.translate(table) for word in tokens]
    sentences = [word.lower() for word in lines]
    
    #print(len(sentences))
    return sentences
```

```python
sentences = prepare_sentences(doc)
print(sentences)
print(sentences[0])
print(len(sentences))
```

```python
def my_function(sentences):
    table = pd.DataFrame(index = range(12), columns = range(11))
    for row in range(12):
        for i in range(len(sentences[:11])): 
            table.iloc[row,i] = sentences[i]
        #[word for word in sentences[0]]
        
    # switch word order
    # sentence 2
    table.iloc[1,5],table.iloc[1,6] = table.iloc[0,7],table.iloc[0,8]
    table.iloc[1,7],table.iloc[1,8] = table.iloc[0,5],table.iloc[0,6]
  
    # sentence 3
    table.iloc[2,3],table.iloc[2,4] = table.iloc[0,5],table.iloc[0,6]
    table.iloc[2,5],table.iloc[2,6] = table.iloc[0,3],table.iloc[0,4]   
    
    # sentence 4
    table.iloc[3,3],table.iloc[3,4] = table.iloc[0,5],table.iloc[0,6]
    table.iloc[3,5],table.iloc[3,6] = table.iloc[0,7],table.iloc[0,8]
    table.iloc[3,7],table.iloc[3,8] = table.iloc[0,3],table.iloc[0,4]
    
    # sentence 5
    table.iloc[4,3],table.iloc[4,4] = table.iloc[0,7],table.iloc[0,8]
    table.iloc[4,5],table.iloc[4,6] = table.iloc[0,3],table.iloc[0,4]
    table.iloc[4,7],table.iloc[4,8] = table.iloc[0,5],table.iloc[0,6] 
    
    # sentence 6
    table.iloc[5,3],table.iloc[5,4] = table.iloc[0,7],table.iloc[0,8]
    table.iloc[5,7],table.iloc[5,8] = table.iloc[0,3],table.iloc[0,4]
    
    print(table)
    return (table)
```

```python
def create_tables(sentences):
    
    tables = {}
    number_sentences = int(len(sentences)/11)
    print(number_sentences)
    
    for k in range(number_sentences):
        tables[k] = pd.DataFrame(index = range(12), columns = range(11))
    
    for k in range(1, number_sentences + 1):
        for row in range(12):
            for i in range(11):
                tables[k - 1].iloc[row,i] = sentences[i + (k-1)*11]   
                
    # switch word order
    for k in range(number_sentences):
        # sentence 2
        tables[k].iloc[1,5],tables[k].iloc[1,6] = tables[k].iloc[0,7], tables[k].iloc[0,8]
        tables[k].iloc[1,7],tables[k].iloc[1,8] = tables[k].iloc[0,5],tables[k].iloc[0,6]

        # sentence 3
        tables[k].iloc[2,3],tables[k].iloc[2,4] = tables[k].iloc[0,5],tables[k].iloc[0,6]
        tables[k].iloc[2,5],tables[k].iloc[2,6] = tables[k].iloc[0,3],tables[k].iloc[0,4]   

        # sentence 4
        tables[k].iloc[3,3],tables[k].iloc[3,4] = tables[k].iloc[0,5],tables[k].iloc[0,6]
        tables[k].iloc[3,5],tables[k].iloc[3,6] = tables[k].iloc[0,7],tables[k].iloc[0,8]
        tables[k].iloc[3,7],tables[k].iloc[3,8] = tables[k].iloc[0,3],tables[k].iloc[0,4]

        # sentence 5
        tables[k].iloc[4,3],tables[k].iloc[4,4] = tables[k].iloc[0,7],tables[k].iloc[0,8]
        tables[k].iloc[4,5],tables[k].iloc[4,6] = tables[k].iloc[0,3],tables[k].iloc[0,4]
        tables[k].iloc[4,7],tables[k].iloc[4,8] = tables[k].iloc[0,5],tables[k].iloc[0,6] 

        # sentence 6
        tables[k].iloc[5,3],tables[k].iloc[5,4] = tables[k].iloc[0,7],tables[k].iloc[0,8]
        tables[k].iloc[5,7],tables[k].iloc[5,8] = tables[k].iloc[0,3],tables[k].iloc[0,4]


    return (tables)
```

```python
tables = create_tables(sentences)
print(tables)
```
